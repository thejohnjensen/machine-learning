```python
# Visualize Crime vs Pupil/Teacher ratio
```

```python
boston.data[0][0]
```

```python
bos = pd.DataFrame(boston.data)
bos.columns = boston.feature_names
```

```python
x = bos['PTRATIO']
y = bos['CRIM']
color = ['blue' if val == 0.0 else 'red' for val in bos['CHAS']] 
```

```python
plt.title('Crime Rate vs Teacher/Pupil Ratio')
plt.ylabel('Crime Rate Per Capita')
plt.xlabel('Pupil/Teacher Ratio \n *Red mark indicates bounded by Charles River.')
plt.scatter(x, y, c=color)
plt.show()
```

```python
bk = np.sqrt(bos['B']/1000) + 0.63
y = bk
x = bos['DIS']
plt.title('Proportion of black citizens vs Distance to Employment Centers')
plt.ylabel('Proportion of black citizens by town.')
plt.xlabel('Distance to employment centers.')
plt.scatter(x, y)
plt.show()
```

```python
y = bos['NOX'] * 10
x = boston.target
z = bos['INDUS']
plt.title('Median Home Value vs Nitric Oxide and Proportion of Non-Retail Businesses')
plt.ylabel('Nitric Oxide PPM (blue) \n Proportion of Non-Retail Businesses (red)')
plt.xlabel('Median Home Value x $1000')
plt.scatter(x, y, c='blue')
plt.scatter(x, z, c='red')
plt.show()
```

```python
y = bos['CRIM'] 
x = boston.target
plt.title('Median Home Value vs Crime Rate ')
plt.ylabel('Crime Rate per Capita')
plt.xlabel('Medain Home Cost')
plt.scatter(x, y)
plt.show()
```
